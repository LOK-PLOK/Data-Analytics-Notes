#### Import Packages

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Numpy (Numerical Python)
NumPy is a library for numerical computing with support for multi-dimensional arrays and mathematical operations.

Key Features:
- `ndarray`: Efficient multi-dimensional arrays.
- Mathematical operations: `sum()`, `mean()`, `std()`, `max()`, `min()`, etc.
- Indexing & Slicing: Access parts of arrays efficiently.
- Broadcasting: Perform operations on arrays of different shapes.

```python
arr = np.array([1, 2, 3, 4, 5])  # Create array  
print(arr.shape)  # Check shape  

# Operations  
arr2 = arr * 2  # Multiply elements  
arr3 = np.sqrt(arr)  # Square root  
```

# Matplotlib (Data Visualization)
Matplotlib is a plotting library for Python, used to create graphs and charts.

âœ… Key Plot Types:
- Line Plot: `plt.plot(x, y)`
- Bar Chart: `plt.bar(categories, values)`
- Histogram: `plt.hist(data, bins=n)`
- Scatter Plot: `plt.scatter(x, y)`

ðŸ”¹ Customization Options:
- Title & Labels: `plt.title()`, `plt.xlabel()`, `plt.ylabel()`
- Legend: `plt.legend()`
- Grid: `plt.grid(True)`
- Colors & Styles: `color`, `linestyle`, `marker`

## Basic Example

```python

x = [1, 2, 3, 4, 5]  
y = [10, 20, 25, 30, 40]  

plt.plot(x, y, marker='o', linestyle='--', color='b', label="Data")  
plt.xlabel("X Axis")  
plt.ylabel("Y Axis")  
plt.title("Line Plot Example")  
plt.legend()  
plt.grid(True)  
plt.show()  
```

## Bar Chart example:

```python
categories = ['A', 'B', 'C', 'D']  
values = [5, 8, 12, 7]  

plt.bar(categories, values, color='g')  
plt.xlabel("Categories")  
plt.ylabel("Values")  
plt.title("Bar Chart Example")  
plt.show()  
```

## Histogram Example


```python

data = np.random.randn(1000)  # Generate random data  
plt.hist(data, bins=30, color='purple', edgecolor='black')  
plt.xlabel("Values")  
plt.ylabel("Frequency")  
plt.title("Histogram Example")  
plt.show()  
```

# Pandas (Data Manipulation & Analysis)
Pandas is a library for handling structured tabular data using DataFrames and Series.

âœ… Key Features:
- `Series`: 1D labeled array.
- `DataFrame`: 2D table (like an Excel sheet).
- Reading/Writing Files: Supports `.csv`, `.xlsx`, `.json`, etc.
- Filtering & Indexing: Select and modify data easily.
- Aggregation: Sum, mean, count, etc.


## Basic Usage

```python
# Create DataFrame  
data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35]}  
df = pd.DataFrame(data)  

print(df)  # Display DataFrame  
print(df['Age'].mean())  # Calculate mean age  
```

## Reading a CSV File:

```python
df = pd.read_csv('data.csv')   # this will error because there is no data
print(df.head())  # Show first 5 rows  
```

## Filtering Data:

```python
filtered_df = df[df['Age'] > 25]  # Get rows where Age > 25  
print(filtered_df)  
```

## Adding a new column:

```python
df['Salary'] = [50000, 60000, 55000]  # Add a new column  
print(df)  
```

## GroupBy & Aggregation:

```python
df.groupby('Category')['Sales'].sum()  # Sum sales by category  
```
